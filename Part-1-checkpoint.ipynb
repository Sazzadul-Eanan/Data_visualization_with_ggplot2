### This part demonstrates the preparation of basic graphs using ggplot2

```r
# call the library
library(ggplot2)
```

```r
# a scatterplot to view the relationship between the variables Sepal.Length and Petal.Length
p <- ggplot(data=iris, aes(y=Sepal.Length, x=Petal.Length))+geom_point()
```

```r
# view the plot (a scatterplot can be drawn among two or, more discrete variables)
p     
```

```r
# tuning more advanced functions of aesthetics
p2 <- ggplot(data=iris, aes(y=Sepal.Length, x=Petal.Length, col=Species))+geom_point()
```

```r
# view the plot according to color variations for individual species
p2
```

```r
# tuning more advanced functions of aesthetics
p3 <- ggplot(data=iris, aes(y=Sepal.Length, x=Petal.Length, col=Species, shape=Species))+geom_point()
```

```r
# view the plot according to color as well as shape variations for individual species
p3
```

```r
# a histogram to view the data distribution of the variable 'Sepal.Width'
q <- ggplot(data=iris, aes(x=Sepal.Width))+geom_histogram()
```

```r
# view the plot (a histogram can be drawn by a continuous variable)
q
```

```r
# reset the bin to have more adjascent view of the bars
q2 <- ggplot(data=iris, aes(x=Sepal.Width))+geom_histogram(bins=15)
```

```r
# view the plot
q2
```

```r
# color the bins
q3 <- ggplot(data=iris, aes(x=Sepal.Width))+geom_histogram(bins=15, fill="blue")
```

```r
# view the plot
q3
```

```r
# color the column boarder
q4 <- ggplot(data=iris, aes(x=Sepal.Width))+geom_histogram(bins=15, fill="blue", col='green')
```

```r
# view the plot
q4
```

```r
# view the Sepal.Width for different species
q5 <- ggplot(data=iris, aes(x=Sepal.Width, fill=Species))+geom_histogram(bins=15)
```

```r
# view the plot
q5
```

```r
# smoothline assuming that the petal-width(in y-axis) is dependent on the sepal-width(in x-axis)
n <- ggplot(data=iris, aes(y=Petal.Width, x=Sepal.Width))+geom_smooth()
```

```r
# view the plot with 95% confidence interval
n
```

```r
# smoothline assuming that the petal-width(in y-axis) is dependent on the sepal-width(in x-axis)
n2 <- ggplot(data=iris, aes(y=Petal.Width, x=Sepal.Width))+geom_smooth(se=F)       # standard error = false
```

```r
# view the plot WITHOUT 95% confidence interval
n2
```

```r
# smoothline assuming that the petal-width(in y-axis) is dependent on the sepal-width(in x-axis) for different Species
n3 <- ggplot(data=iris, aes(y=Petal.Width, x=Sepal.Width, col=Species))+geom_smooth()
```

```r
n3
```

```r
# smoothline assuming that the petal-width(in y-axis) is dependent on the sepal-width(in x-axis) with customized line method
n4 <- ggplot(data=iris, aes(y=Petal.Width, x=Sepal.Width, col=Species))+geom_smooth(method=lm, se=F)+geom_point()  
```

```r
# view the plot
n4
```

```r
# smoothline assuming that the petal-width(in y-axis) is dependent on the sepal-width(in x-axis) with customized line method
# and faceting the identical groups according to the variable 'Species'
n5 <- ggplot(data=iris, aes(y=Petal.Width, x=Sepal.Width, col=Species))+geom_point()+geom_smooth(method=lm, se=F)+facet_grid(Species)
```

```r
# view the plot
n5
```

### Working with graph theme

```r
# Add title to the graph and reanming the y-axis
E = e3 + labs(title="Boxplot of Lung Capacity", y="Lung Capacity")     # e3 contains the previous code
```

```r
# view the plot
E
```

```r
# changing plot background
EE = E + theme(panel.background =element_rect("palegreen2"))
```

```r
# view the plot
EE
```

```r
# centering and editing the plot title
EEE = EE + theme(plot.title=element_text(hjust=0.5, face="bold", colour="red"))      # hjust = horizontal justification
```

```r
# view the plot
EEE
```
